```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
import io
import numpy as np
from datetime import datetime

# read file
url = "https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv"
df = pd.read_csv(url)
df.columns = df.columns.str.strip().str.replace('"', '')

# Print the headingd and the first two rows
print(df.head(2))

# Store the colum names as a list
column_names_list = df.columns.tolist()
print(column_names_list)

# Convert two columns to numpy arrays
height_array = df['Height(Inches)'].to_numpy()
weight_array = df['Weight(Pounds)'].to_numpy()

# Display data as a scatter plot graph
plt.figure(figsize=(10, 6))
plt.scatter(height_array, weight_array)
plt.xlabel('Height(Inches)')
plt.ylabel('Weight(Pounds)')
plt.title('Height vs Weight')
plt.show()
```
